```python
# some data to play with
x1 = range(0,300)
y1 = [1/(n+1) for n in x]

x2 = range(0,300)
y2 = [2 ** n for n in x]

x3 = range(0,300)
y3 = [n ** 3 for n in x]

x4 = range(0,300)
y4 = np.sqrt(x)

# plot the first subplot
plt.subplot(224)
plt.plot(x1, y1)
plt.title('1/(x+1)')

# the second subplot
plt.subplot(223)
plt.plot(x2, y2)
plt.title('2 ** x')

plt.subplot(222)
plt.plot(x3, y3)
plt.title('x ** 3')

plt.subplot(221)
plt.plot(x4, y4)
plt.title('sqrt(x)')

plt.suptitle('Subplots')
plt.tight_layout()

plt.show()
```
