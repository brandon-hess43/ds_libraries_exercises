## 4. Load the chipotle dataset from SQL, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
import env
```

```python
def get_db_url(user, password, host, database):
    return f'mysql+pymysql://{user}:{password}@{host}/{database}'
```

```python
url = get_db_url(env.user, env.password, env.host, 'chipotle')
```

```python
chipotle = pd.read_sql('select * from orders', url)
chipotle
```

```python
chipotle.item_price 
```

```python
chipotle.price = chipotle.item_price.str.replace('$', '').str.replace(',','').astype('float')
chipotle.price.dtypes
```

```python
chipotle
```

```python
top_four = chipotle.groupby('item_name').price.sum().sort_values(ascending=False).head(4)
top_four
```

```python
top_four=top_four.reset_index()
top_four
```

```python
sns.barplot(top_four,x='item_name',y='price')
```
